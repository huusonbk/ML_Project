```python
#In general, we can download the .csv file from web brower to input into notebook, but it's preferable to create a small function to fetch latest data.
#We can also set up a scheduled job to do that automatically at regular intervals.
import os
import tarfile
import urllib
import urllib.request
Download_root = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
Housing_path = os.path.join("datasets", "housing")
Housing_url = Download_root + "datasets/housing/housing.tgz"
def fetch_housing_data(housing_url = Housing_url, housing_path = Housing_path):
  os.makedirs(housing_path,exist_ok= True)
  tgz_path = os.path.join(housing_path, "housing.tgz")
  urllib.request.urlretrieve(housing_url, tgz_path)
  housing_tgz = tarfile.open(tgz_path)
  housing_tgz.extractall(path=housing_path)
  housing_tgz.close()
```

```python
# Test features whether follow normal distribution D’Agostino’s K² Test & assume H0: Normal Distribution, H1: Not distribution
import numpy as np
from scipy.stats import normaltest
def check_normal_dist_feature(df):
  df = df.fillna(0)
  df_P_value = pd.DataFrame()
  for feature in df.columns:
    data =df[feature].values
    try: 
      # normality test
      stat, p = normaltest(data)
      alpha = 0.05
      if p > alpha:
        conclude = "Gaussian"
      else:
        conclude = 'not look Gaussian'
      df_conclude_each_feature= pd.DataFrame({"feature":[feature], "p_value": [p], "conclude":[conclude] })
      df_P_value = df_P_value.append(df_conclude_each_feature)
    except:
      continue
  df_P_value.reset_index(drop=True, inplace=True)
  return df_P_value
```

* When we split the data,we need to carefully consider the significant sampling bias.

##Customing Transformers



*   Although Scikit-learn provides many available useful transformers, we need to write the task such as cleanup operations or combining specific attributes.
*  If we want our Transformers work seamlessly with scikit-learn function, we need to create a class and implement with 3 methods: fit(), Transform(), fit_transform().
* We can get the last one for free by simply adđing TransformerMixin as a base class.
* We will also two extra methods (get_params() and set_params) that will be useful for automatic hyperparameter tuning.



```python
from sklearn.base import  BaseEstimator, TransformerMixin

room_id, bedroom_id, population_id, households_id = 3,4,5,6

class CombineAttributeAdder(BaseEstimator, TransformerMixin):
  def __init__(self, add_bedrooms_per_room = True):
    self.add_bedrooms_per_room = add_bedrooms_per_room
  def fit(self,X, y = None):
    return self
  def transform(self,X):
    rooms_per_household = X[:, room_id]/X[:, households_id]
    population_per_household = X[:, population_id]/X[:, households_id]
    if self.add_bedrooms_per_room:
      bedrooms_per_room = X[:, bedroom_id]/X[:, room_id]
      return np.c_[X,rooms_per_household,population_per_household,  bedrooms_per_room]
    else:
      return np.c_[X,rooms_per_household,population_per_household]
```

```python
attribute_adder = CombineAttributeAdder(add_bedrooms_per_room=False)
Housing_extra_att = attribute_adder.transform(housing.values)
```

##Feature Scaling

There are 2 types for feature scaling:


1.   Min-max scaling (Normalization): Value is scaled range from [0,1] by substracting the min value and dividing by the max minus the min. This method is sensitve with outliers.
2. Standardization: Substract values with the mean values and then dividing by the standard deviation. There are no bound values for this method, and it's not suitable for neural network.    

We can combine this step in to next step called: Transformation Pipelines


##Transformation pipeline

There are many data transformation steps that need to be executed in the right order, we can use the Scikit-learn Pipeline class to help with the sequences of transformation.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
                         ("Missing_data_filling", SimpleImputer(strategy = "median")),
                         ("Attribue_adder", CombineAttributeAdder()),
                         ("Standarization", StandardScaler())
])

housing_num_tr = num_pipeline.fit_transform(housing_num)

#We can use the ColumnTransformer to transform each Column in Pandas DataFrame,  it's suitble for combine the Numerical data and Categorical data
from sklearn.compose import ColumnTransformer

num_attribute = list(housing_num)
cat_attribute = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
                                  ("Num", num_pipeline, num_attribute),
                                   ("Cat", OneHotEncoder(), cat_attribute)
])

housing_prepared = full_pipeline.fit_transform(housing)
```

#Step 5: Select and Train model

##Training and Evaluating on the Training Set

```python
#Trying to employ the Linear Regression model to fit the model

from sklearn.ensemble import RandomForestRegressor
RF_reg = RandomForestRegressor()
RF_reg.fit(housing_prepared, housing_target)
#Evaluate it on the training set
from sklearn.metrics import mean_squared_error
Housing_prediction = RF_reg.predict(housing_prepared)
RF_mse = mean_squared_error(housing_target,Housing_prediction)
RF_rmse = np.sqrt(RF_mse)
```

```python
print(RF_rmse)
```

When we train and evaluate on training set, the performance of training set is really good, but we are likely to be traped in the overfiting of model on training dataset. To ensure that our model whether overfit or not, we can evaluate model on testing dataset

##Better Evaluation using K-fold cross valiadation



1.   We can use the Train_test_split() to split the training dataset into small training dataset and validation dataset to test model.
2.   Another method can be used is K-fold Cross validation, we can split the training data set k subsets, pick k-1 subsets for training and the remaining for validating model, repeat this process for k time. The training set and validation during cross-validation is distinct.



```python
from sklearn.model_selection import cross_val_score
rf_score = cross_val_score(RF_reg, housing_prepared, housing_target, scoring = "neg_mean_squared_error", cv = 10)
rf_rmse_scores = np.sqrt(-rf_score)
```

```python
#Display scores
def display_scores(Score):
  print("Score is here:", Score)
  print("Mean of Score is here:", Score.mean())
  print("Standard deviation of Score is here:", Score.std())
display_scores(rf_rmse_scores)
```

#Step 6: Fine_Tune model

##Grid Search



1.   Normally, we will change the hyperparameter to find the best performance of model.
2.   Changing manually the hyperparamter is time-consumming, it's not suitable with the model with many hyperparameter.
3. One method called grid search to implement all of hyperparameter that you want to experiment with and what values to try out.This method also uses the Cross-Validation for model evaluation



```python
from sklearn.model_selection import GridSearchCV

Param_grid = [
              {'n_estimators':[3,10,30], 'max_features': [2,4,6,8]},
              {'bootstrap': [False], 'n_estimators':[3,10], 'max_features': [2,3,4]}
              ]

forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, Param_grid, cv = 5, scoring = 'neg_mean_squared_error', return_train_score = True)
grid_search.fit(housing_prepared, housing_target)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
#Evaluate score for all hyperparameter
Result = grid_search.cv_results_
for mean_score, params in zip(Result['mean_test_score'], Result['params']):
  print(np.sqrt(-mean_score),params )
```
